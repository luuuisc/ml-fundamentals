### Exploración de datos 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el archivo CSV
df = pd.read_csv('data.csv')

# Mostrar las primeras filas del DataFrame
df.head(9)  
```

```python
# Estadísticas descriptivas generales
df.describe()
```

```python
# Promedio de goles como local y visitante
goles_local_cebo = df[df['equipo_local'] == 'Cebollitas FC']['goles_local'].mean()
goles_visitante_cebo = df[df['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'].mean()

print(f"Goles promedio de Cebollitas FC como local: {goles_local_cebo}")
print(f"Goles promedio de Cebollitas FC como visitante: {goles_visitante_cebo}")
```

```python
# Histograma de goles marcados (local y visitante)
fig, ax = plt.subplots(1,2, figsize=(12, 4))

sns.histplot(df['goles_local'], bins=10, kde=True, ax=ax[0])
ax[0].set_title('Distribución de Goles Locales')

sns.histplot(df['goles_visitante'], bins=10, kde=True, ax=ax[1], color='orange')
ax[1].set_title('Distribución de Goles Visitantes')

plt.tight_layout()
plt.show()
```

```python
# Boxplot para goles del equipo local y visitante
sns.boxplot(x=df['goles_local'])
plt.title('Boxplot de Goles Locales')
plt.show()

sns.boxplot(x=df['goles_visitante'], color='orange')
plt.title('Boxplot de Goles Visitantes')
plt.show()
```

```python
# Scatterplot de posesión vs goles marcados (local y visitante)
sns.scatterplot(x='posesion_local (%)', y='goles_local', data=df)
plt.title('Relación posesión vs Goles marcados (equipo local)')
plt.xlabel('Posesión Local (%)')
plt.ylabel('Goles Locales')
plt.show()

sns.scatterplot(x='posesion_visitante (%)', y='goles_visitante', data=df, color='orange')
plt.title('Relación posesión vs Goles marcados (equipo visitante)')
plt.xlabel('Posesión Visitante (%)')
plt.ylabel('Goles Visitantes')
plt.show()
```

```python
# Mapa de calor de correlación entre variables clave
plt.figure(figsize=(10, 6))
sns.heatmap(df[['goles_local', 'goles_visitante', 'posesion_local (%)', 
                'posesion_visitante (%)', 'tiros_arco_local',
                'tiros_arco_visitante']].corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlación')
plt.show()
```
